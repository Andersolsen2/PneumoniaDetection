```python
import os 
import numpy as np 
import pandas as pd
from utilities import *


cwd = os.getcwd()
print (cwd)


## Read labels
true_labels=pd.read_csv('./stage_2_train_labels.csv')
pred_labels_070=pd.read_csv('./submission1_conf=0.7.csv')
pred_labels_060=pd.read_csv('./submission1_conf=0.6.csv')
pred_labels_050=pd.read_csv('./submission1_conf=0.5.csv')
pred_labels_040=pd.read_csv('./submission1_conf=0.4.csv')

## suidID

ID_true = true_labels['patientId']
ID_pred_070 = pred_labels_070['patientId']
ID_pred_060 = pred_labels_060['patientId']
ID_pred_050 = pred_labels_050['patientId']
ID_pred_040 = pred_labels_040['patientId']



## PARSE
# Test labels
true_parsed = test(true_labels,ID_true,ID_pred_050)

# Predicted
pred_parsed_070 = predicted(pred_labels_070, ID_pred_070)
pred_parsed_060 = predicted(pred_labels_060, ID_pred_060)
pred_parsed_050 = predicted(pred_labels_050, ID_pred_050)
pred_parsed_040 = predicted(pred_labels_040, ID_pred_040)


## Calculate true positiv, true negativ, false positiv, etc. 
#  IOU treshhold
iou = 0.5
sum_of_all, True_positiv_070, True_negativ_070, False_positiv_070, False_negativ_070 = calucalteTP_TN_FP_FN(pred_parsed_070, true_parsed, iou)
_, True_positiv_060, True_negativ_060, False_positiv_060, False_negativ_060 = calucalteTP_TN_FP_FN(pred_parsed_060, true_parsed, iou)
_, True_positiv_050, True_negativ_050, False_positiv_050, False_negativ_050 = calucalteTP_TN_FP_FN(pred_parsed_050, true_parsed, iou)
_, True_positiv_040, True_negativ_040, False_positiv_040, False_negativ_040 = calucalteTP_TN_FP_FN(pred_parsed_040, true_parsed, iou)



## PLOT

import matplotlib.pyplot as plt

# conf 0.50

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["True Positive: "+str(True_positiv_050),
          "True Negative: "+str(True_negativ_050),
          "False Positive: "+str(False_positiv_050),
          "False Negative: "+str(False_negativ_050)]

data = [True_positiv_050, True_negativ_050, False_positiv_050, False_negativ_050]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-40)

bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(xycoords='data', textcoords='data', arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                 horizontalalignment=horizontalalignment, **kw)

ax.set_title("Results with a confidence above 0.50")
plt.savefig('piechart_050_CHex.png')
plt.show()


## IOU PLOT

tp_50 = []
tn_50 = []
fp_50 = []
fn_50 = []

tp_60 = []
tn_60 = []
fp_60 = []
fn_60 = []


iou_steps = np.linspace(0.1,0.9,9)
for i in range(0,len(iou_steps),1):
    iou=iou_steps[i,]
    print(iou)
    _, True_positiv_060, True_negativ_060, False_positiv_060, False_negativ_060 = calucalteTP_TN_FP_FN(pred_parsed_060, true_parsed, iou)
    _, True_positiv_050, True_negativ_050, False_positiv_050, False_negativ_050 = calucalteTP_TN_FP_FN(pred_parsed_050, true_parsed, iou)
    
    
    tp_50.append(True_positiv_050)
    tn_50.append(True_negativ_050)
    fp_50.append(False_positiv_050)
    fn_50.append(False_negativ_050)
    
    tp_60.append(True_positiv_060)
    tn_60.append(True_negativ_060)
    fp_60.append(False_positiv_060)
    fn_60.append(False_negativ_060)
    
    
    
plt.plot(iou_steps, tp_50 , 'r--', iou_steps, tp_60, 'b--')
plt.plot(iou_steps, fp_50 , 'r-.', iou_steps, fp_60, 'b-.')
plt.ylabel('Number of patients [N]')
plt.xlabel('Intersection over union (IOU)')
plt.legend(['TP w. conf. 0.50','TP w. conf. 0.60','FP w. conf. 0.50','FP w. conf. 0.60'])
plt.savefig('changeIOUCheX.png')
plt.show()
    





    
```
